$$\begin{equation*}dt = t_{or} + wt + ut - tP\end{equation*}$$

$$\begin{equation*}tZP = \frac{aFP}{DR}\end{equation*}$$

## **MODEL**

* **Sets**:
    * $F: {1,2,...nF}$ (Food Banks) 
    * $P: {1,2,...nP}$ (PODs - Points of Distribution)
    * $Z: {1,2,...nZ}$ (Zone centroides)
* **Indeces**:
    * $i: F$
    * $j: P$
    * $k: Z$
* **Parameters**:
    * $pZ_k$: Population in $Z_k$
    * $dZP_{k,j}$: Distance from $Z_k$ centroid to $P_j$ (Miles)
    * $dFP_{i,j}$: Distance from $F_i$ to $P_j$ (Miles)
    * $cF_i$: Food Bank $F_i$ capacity (Pounds)
    * $cP_j$: POD $P_j$ capacity (Pounds)
    * $LCM$: Logistic Cost per Mile (Units?)
    * $LCH$: Logistic Cost per Hour (Units?)
    * $AFP$: amount of Food required by Person (Pounds/Hour)
    * $nTF$: Trucks available in each Food Bank $F_j$
    * $TC$: Trucks' Capacity (Pounds)
    * $TS$: Trucks' Speed (Miles/Hour)
    * $WS$: Walking Speed (Miles/Hour)
    * $UR$: Unloading Rate (Hours/Pound)
    * $PF$: % of People requiring Food
    * $T$: Period planning time (Hours)
    * $tt_{i,j}$: transportation time (hours) it takes to deliver food from $F_i$ to $P_j$
    * $wt_{k,j}$: walking time (hours) that people in $Z_k$ need to reach $P_j$ location
    
* **Decision Variables**:
    * $$
        x_{i,j} : \begin{cases}
            1 & \text{if $F_i$ sends food to $P_j$} \\
            0 & \text{otherwise}
        \end{cases}
        $$
    * $$
        y_{k,j} : \begin{cases}
            1 & \text{if $Z_k$ is served by $P_j$} \\
            0 & \text{otherwise}
        \end{cases}
        $$
    
    * $aFP_{i,j}$: amount of food (pounds) sent from $F_i$ to $P_j$
    * $aZP_{k,j}$: amount of food (pounds) that $Z_k$ require from $P_j$
    * $ort_{j}$: Order received time (hours)
    * $opt_{j}$: Order placed time (hours)
    * $tP_{j}$: time when the amount of food in POD $P_j$ is zero.
    * $dt_{k,j}$: deprivation time (hours) that people in $Z_k$ wait while food is available in $P_j$. Por proporcion de la poblacion servida en ese POD.
    * $fFP_{i,j}$: delivery frequency of trucks sent from $F_i$ to $P_j$
    * $ut_{i,j}$: unloading time (hours) in $P_j$

* **Constraints**:
    
    * **Relation between $F_i$, $P_j$, $Z_k$**:
        * A Food Bank $F_i$ sends food to at least one POD $P_j$:
            $$\begin{equation}\sum_{j \in P}^{}x_{i,j} \geq 1 \quad \forall \quad i \in F\end{equation}$$

        * A POD $P_j$ receives food only for one Food Bank $F_i$:
            $$\begin{equation}\sum_{i \in F}^{}x_{i,j} = 1 \quad \forall \quad j \in P\end{equation}$$

        * A Zone $Z_k$ is served by at least one POD $P_j$:
            $$\begin{equation}\sum_{j \in P}^{}y_{k,j} \geq 1 \quad \forall \quad k \in Z\end{equation}$$

        * A POD $P_j$ serves food only to one Zone $Z_k$:
            $$\begin{equation}\sum_{k \in Z}^{}y_{k,j} = 1 \quad \forall \quad j \in P\end{equation}$$
    
    * **Food capacity**:
        * The amount of food sent from $F_i$ to $P_j$ is related with the capacity of the trucks:
            $$\begin{equation}aFP_{i,j} \leq fFP_{i,j}*TC\end{equation}$$

        * The amount of food sent from $F_i$ to $P_j$ cannot exceed the capacity of the Food Bank:
            $$\begin{equation}\sum_{j \in P}^{}aFP_{i,j} \leq cF_j \quad \forall \quad i \in F\end{equation}$$
        
        * The amount of food sent from $F_i$ to $P_j$ cannot exceed the capacity of the POD:
            $$\begin{equation}\sum_{i \in P}^{}aFP_{i,j} \leq cP_j \quad \forall \quad j \in P\end{equation}$$

        * The amount of food that $Z_k$ gets from $P_j$ must be equal to the food that $F_i$ sends to $P_j$:
            $$\begin{equation}\sum_{i \in F}^{}aFP_{i,j} = \sum_{k \in Z}^{}aZP_{k,j} \quad \forall \quad j \in P\end{equation}$$
        
        * The amount of food required in a period of time $T$ must satisfy the demand:
            $$\begin{equation}\sum_{j \in P}^{}aZP_{k,j}*fFP_{i,j} \leq AFP*PF*pZ_k*T \quad \forall \quad i \in F \quad k \in Z\end{equation}$$

    * **Times**:
        * Each Zone (People) requires a walking time (hours) from its location $Z_k$ to the POD's location $P_j$ to get food:

            **Walking Time** &rarr; 
            $$\begin{equation}wt_{k,j} = \frac{dZP_{k,j}}{WS}*y_{k,j} \quad \forall \quad k \in Z, \quad j \in P \end{equation}$$

        * Each POD $P_j$ requires a time (hours) for unloading the food:

            **Unloading Time** &rarr; 
            $$\begin{equation}ut_{j} = UR*aFP_{i,j}  \quad \forall \quad i \in F, \quad j \in P\end{equation}$$

        * Each truck requires a time (hours) for transportation the food from $F_i$ to $P_j$:

            **Transportation Time** &rarr; $\times 1$ in Deprivations Cost, $\times 2$ in Logistic Cost &rarr;
            $$\begin{equation}tt_{i,j} = \frac{dFP_{i,j}}{TS}*x_{i,j}  \quad \forall \quad i \in F, \quad j \in P\end{equation}$$

        * Each Zone $Z_k$ has asociated a deprivation time related to the POD $P_j$ which served them:
            $$\begin{equation}dt_{k,j} +  tP_j = ort_j + ut_{i,j} + wt_{k,j} \quad \forall \quad k \in Z, \quad j \in P\end{equation}$$

            $$\begin{equation}tP_j = ut_{i,j} + wt_{k,j} + \frac{aZP_{k,j}}{AFP*PF*pZ_k} \quad \forall \quad k \in Z, \quad j \in P\end{equation}$$

            $$\begin{equation}opt_j + tt_{i,j} \leq ort_j \quad \forall \quad i \in F, \quad j \in P\end{equation}$$

            $$\begin{equation}ort_j \geq 24 (hours) \quad \forall \quad j \in P\end{equation}$$
            
            $$\begin{equation}ort_j*fFP_ij <= 24*T \quad \forall \quad i \in F \quad j \in P \end{equation}$$

**Objective Function**

* $SC$: Social Cost
* $DC$: Deprivation Cost
* $LC$: Logistic Cost
* $TC$: Transportation Cost
* $UC$: Unloading Cost

$$\begin{equation*}\min SC\end{equation*}$$

$$\begin{equation*}SC = DC + LC\end{equation*}$$

$$\begin{equation*}DC_{k,j} = \frac{3066*dt_{k,j}+349.66*dt_{k,j}^2}{ER_{\$COP \rightarrow \$US}} \quad \forall \quad k \in Z\end{equation*}$$

$$\begin{equation*}dt_{k,j} = tOR_{i,j} + wt_{k,j} + ut_{j} - tP_{j}\end{equation*}$$

$$\begin{equation*}LC = TC + UC\end{equation*}$$

$$\begin{equation*}TC = \frac{aFP_{i,j}}{cT}*2*dFP_{i.j}*\left[LCM + \frac{LCH}{sT}\right]\end{equation*}$$

$$\begin{equation*}UC = UR*aFP_{i,j}*LCH\end{equation*}$$

$$\begin{equation*}\min SC\end{equation*}$$

```python
import gurobipy as gp

# **Sets**:
nF = 1
F = np.arange(1, nF+1) #(Food Banks) i
nP = 228
nZ = 96

# **Parameters**:
pZ = pd.read_excel(pd.ExcelFile(data_path), sheet_name='ZIP_CODE_POPULATION-DEMAND', header=None)
pZ_k = pZ.iloc[1:,1:2].astype(float) # Population in Zone $Z_k$ (People) pZ_k.loc[1].values[0]
pZ_k_dict = pZ_k[1].to_dict()

dPZ_file = pd.read_excel(pd.ExcelFile(data_path), sheet_name='DISTANCE_POD_ZIP', index_col=0)
dPZ_jk = dPZ_file.astype(float) # Distance from $Z_k$ centroid to $P_j$ (Miles) dZP_kj.loc[1,1]
dPZ_jk_dict = dPZ_jk.T.to_dict()

dFP = pd.read_excel(pd.ExcelFile(data_path), sheet_name='DISTANCE_FOOD_BANK-PODS', index_col=0)
dFP_ij = dFP.astype(float) #Distance from $F_i$ to $P_j$ (Miles)
dFP_ij_dict = dFP_ij.T.to_dict()

cP = pd.read_excel(pd.ExcelFile(data_path), sheet_name='CAPACITY_POD', index_col=0)
cP_j = cP.iloc[0:,0:1].astype(float) # POD $P_j$ capacity (Pounds)
cP_j_dict = cP_j['Capacity_pounds_of_food_per_day'].to_dict()

# Number of Trucks availables
NT = nP*100
NT_i = pd.DataFrame([NT]) 
NT_i.columns = NT_i.columns + 1 
NT_i.index = NT_i.index + 1
NT_i_dict = NT_i[1].to_dict()

LCM = 0.969 #0.58 # Logistic Cost per Mile ($/Mile)
LCH = 71.78 # Logistic Cost per Hour ($/Hour)
AFP = 13.39/24 # Amount of Food required by Person (Pounds/Hour)

TC = 10000 # Trucks' Capacity (Pounds)
TS = 30 #12.5 # Trucks' Speed (Miles/Hour)
WS = 12.5 #3.1 # Walking Speed (Miles/Hour)
UR = 0.00005 # Unloading Rate (Hours/Pound)
SR = 0.0003 # Setting-up Rate (Hours/Pound)
LR = 0.00005 # Loading Rate (Hours/Pound)

PF = 0.05#0.1 # 0.005#% of People requiring Food
T = 10 # Period planning time (Days)

# Food Bank $F_i$ capacity (Pounds) Big-M
CF = np.sum(AFP*PF*pZ_k, axis=0)*T*24#*100
cF_i = pd.DataFrame([CF])
cF_i.index = cF_i.index + 1
cF_i_dict = cF_i[1].to_dict()

# Walking time (hours) from $Z_k$ to $P_j$
wt_jk = dPZ_jk/WS
wt_jk_dict = wt_jk.T.to_dict()
# Transportation time (hours) from $F_i$ to $P_j$
tt_ij = dFP_ij/TS
tt_ij_dict = tt_ij.T.to_dict()

M_Z = AFP*PF*np.max(pZ_k)*T*24*100
M_T = T*24*100
T_hours = T*24
AFP_PF_pZ_k = AFP*PF*pZ_k
AFP_PF_pZ_k_dict = AFP_PF_pZ_k[1].to_dict()

#PODs = {'PSN': P_small_near, 'PSF': P_small_far}
#Zones = {'ZSN': Z_small_near, 'ZSF': Z_small_far}
PODs = {'PMN': P_medium_near, 'PMF': P_medium_far}
Zones = {'ZMN': Z_medium_near, 'ZMF': Z_medium_far}
#nP = 228
#nZ = 96
#PODs = {'FULL': np.arange(1, nP+1)}
#Zones = {'NETWORK': np.arange(1, nZ+1)}

for keyP, P in PODs.items():
    for keyZ, Z in Zones.items():
        # ##MODEL##
        model = gp.Model('Food_Bank') 

        # ##DECISION VARIABLES##:
        # x_ij if $F_i$ sends food to $P_j$
        x_ij = {(i, j): model.addVar(vtype = gp.GRB.BINARY, name = "x_%d_%d" %(i, j)) for i in F for j in P}
        # y_kj if $Z_k$ is served by $P_j$  
        y_jk = {(j, k): model.addVar(vtype = gp.GRB.BINARY, name = "y_%d_%d" %(j, k)) for j in P for k in Z}

        # Amount of food (pounds) sent from $F_i$ to $P_j$
        aFP_ij = {(i, j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "aFP_%d_%d" %(i, j)) for i in F for j in P} 
        # Amount of food (pounds) that $P_j$ can serve to $Z_k$
        aPZ_jk = {(j, k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "aPZ_%d_%d" %(j, k)) for k in Z for j in P}
        # Number of trucks sent from $F_i$ to $P_j$
        nT_ij = {(i, j): model.addVar(vtype = gp.GRB.INTEGER, name = "nT_%d_%d" %(i, j)) for i in F for j in P} 
        # Order received time (hours) in $P_j$
        ort_ij = {(i,j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "ort_%d_%d" %(i,j)) for i in F for j in P}
        # Order placed time (hours) in $P_j$
        opt_ij = {(i,j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "opt_%d_%d" %(i,j)) for i in F for j in P}
        # Deprivation time (hours) that people in $Z_k$ wait while food is available in $P_j$. Por proporcion de la poblacion servida en ese POD.
        dt_jk = {(j, k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_%d_%d" %(j, k)) for j in P for k in Z}
        # Unloading time (hours) in $P_j$ related to the amount of food received from $F_i$
        ut_ij = {(i, j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "ut_%d_%d" %(i, j)) for i in F for j in P}
        # Delivery frequency of trucks sent from $F_i$ to $P_j$
        fFP_ij = {(i, j): model.addVar(vtype = gp.GRB.INTEGER, name = "fFP_%d_%d" %(i, j)) for i in F for j in P}

        dt_k = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_%d" %(k)) for k in Z}
        dtt_k = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dtt_%d" %(k)) for k in Z}
        dt_k_last = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_last_%d" %(k)) for k in Z}
        #dt_k_start = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_start_%d" %(k)) for k in Z}
        ort_jk = {(j,k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "ort_%d_%d" %(j,k)) for j in P for k in Z}

        dep_cost = model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dep_cost")
        log_cost = model.addVar(vtype = gp.GRB.CONTINUOUS, name = "log_cost")
        
        #Update model (it actully puts everything togeter)
        model.update()

        # ##CONSTRAINTS##:

        Fi_sendsFood_Pj = {model.addConstr(gp.quicksum(x_ij[i,j] for j in P) >= 1, name=f'F{i}_sendsFood_Pj') for i in F}
        Pj_receiveFood_Fi = {model.addConstr(gp.quicksum(x_ij[i,j] for i in F) <= 1, name=f'P{j}_receiveFood_Fi') for j in P}
        Pj_sendsFood_Zk = {model.addConstr(gp.quicksum(y_jk[j,k] for k in Z) <= 1, name=f'P{j}_sendsFood_Zk') for j in P}
        Zk_receiveFood_Pj = {model.addConstr(gp.quicksum(y_jk[j,k] for j in P) >= 1, name=f'Z{k}_receiveFood_Pj') for k in Z}

        Fi_sendsFood_Pj_TruckCapacity = {model.addConstr(aFP_ij[i,j] <= x_ij[i,j]*cP_j_dict[j]) for i in F for j in P}
        FoodBankCapacity = {model.addConstr(gp.quicksum(aFP_ij[i,j] for j in P) <= cF_i_dict[i]) for i in F}
        Truck_available = {model.addConstr(gp.quicksum(nT_ij[i,j] for j in P) <= NT_i_dict[i]) for i in F}

        FoodGet_FoodSent = {model.addConstr(gp.quicksum(aFP_ij[i,j] for i in F) == gp.quicksum(aPZ_jk[j,k] for k in Z)) for j in P} 
        Food_sent_capacity = {model.addConstr(nT_ij[i,j]*TC >= aFP_ij[i,j]) for i in F for j in P}

        UnloadingTime = {model.addConstr(ut_ij[i, j] == (UR+SR)*aFP_ij[i, j]) for i in F for j in P}

        OrderPlacedTime = {model.addConstr(opt_ij[i,j] + 2*tt_ij_dict[i][j]*x_ij[i,j] + LR*ut_ij[i,j] == ort_ij[i,j]) for i in F for j in P}

        FoodDemand = {model.addConstr(gp.quicksum(aPZ_jk[j,k]*fFP_ij[i,j] for i in F for j in P) <= AFP_PF_pZ_k_dict[k]*T_hours) for k in Z}
        DeprivationTime = {model.addConstr(AFP_PF_pZ_k_dict[k]*ort_ij[i,j]*y_jk[j,k] == AFP_PF_pZ_k_dict[k]*dt_jk[j,k] + aPZ_jk[j,k]) for i in F for j in P for k in Z }

        #DeprivationTime2 = {model.addConstr(dt_jk[j,k] >= ut_ij[i,j]*y_jk[j,k] + wt_jk_dict[j][k]*y_jk[j,k]) for i in F for j in P for k in Z }
        #DeprivationTimeStart = {model.addConstr(dt_k_start[k] >= gp.quicksum(ut_ij[i,j]*y_jk[j,k] for i in F for j in P) + wt_jk_dict[j][k]*y_jk[j,k]) for j in P for k in Z }

        PlaningTime = {model.addConstr(fFP_ij[i,j]*ort_ij[i,j] <= T_hours*x_ij[i,j]) for i in F for j in P}

        M_T = T_hours
        #new0 = {model.addConstr(dt_jk[j,k] <= M_T*y_jk[j,k]) for j in P for k in Z }
        new0 = {model.addConstr(dt_jk[j,k] <= ort_ij[i,j]*y_jk[j,k]) for i in F for j in P for k in Z }
        new1 = {model.addConstr(aPZ_jk[j,k] <= cP_j_dict[j]*y_jk[j,k]) for j in P for k in Z}
        new2 = {model.addConstr(x_ij[i,j] <= gp.quicksum(y_jk[j,k] for k in Z)) for i in F for j in P}
        #new3 = {model.addConstr(ut_ij[i,j] <= M_T*x_ij[i,j]) for i in F for j in P}
        new3 = {model.addConstr(ut_ij[i,j] <= ort_ij[i,j]) for i in F for j in P}
        new4 = {model.addConstr(dt_k_last[k] <= M_T) for k in Z}
        #new41 = {model.addConstr(dt_k_start[k] <= M_T) for k in Z}
        new5 = {model.addConstr(dt_k[k] <= M_T) for k in Z}
        new7 = {model.addConstr(ort_ij[i,j] <= M_T*x_ij[i,j]) for i in F for j in P}
        new8 = {model.addConstr(fFP_ij[i,j] <= 10*M_T*nT_ij[i,j]) for i in F for j in P}
        neww81 = {model.addConstr(fFP_ij[i,j] >= x_ij[i,j]) for i in F for j in P}
        #new9 = {model.addConstr(opt_ij[i,j] <= M_T*x_ij[i,j]) for i in F for j in P}
        new9 = {model.addConstr(opt_ij[i,j] <= ort_ij[i,j]) for i in F for j in P}
        new10 = {model.addConstr(nT_ij[i,j] <= NT_i_dict[i]*x_ij[i,j]) for i in F for j in P}#NT_i_dict[i]

        ort_jk_const = {model.addConstr(ort_jk[j,k] == gp.quicksum(ort_ij[i,j]*y_jk[j,k] for i in F)) for j in P for k in Z}
        #DeprivationTimeFinal = {model.addConstr(dt_k_last[k] >= T_hours*y_jk[j,k] - fFP_ij[i,j]*ort_ij[i,j]) for i in F for j in P for k in Z}
        DeprivationTimeFinal = {model.addConstr(AFP_PF_pZ_k_dict[k]*dt_k_last[k] >=  AFP_PF_pZ_k_dict[k]*T_hours*y_jk[j,k] -  AFP_PF_pZ_k_dict[k]*fFP_ij[i,j]*ort_jk[j,k] -  AFP_PF_pZ_k_dict[k]*dt_jk[j,k] - aPZ_jk[j,k]) for i in F for j in P for k in Z}
        dt_k_total = {model.addConstr(dt_k[k] >= dt_jk[j,k] + wt_jk_dict[j][k]*y_jk[j,k]) for j in P for k in Z}
        dtt_k_total = {model.addConstr(dtt_k[k] >= dt_k[k]*fFP_ij[i,j]) for i in F for j in P for k in Z}

        # ##OBJECTIVE FUNCTION##:
        ER = 1/2000 # Exchange Rate USD to COP Base on the paper Discrete choice Cantillo et al. (2018)
        dFP_ij_LCM_LCH_TS = 2*dFP_ij*(LCM+(LCH/TS))
        dFP_ij_LCM_LCH_TS_dict = dFP_ij_LCM_LCH_TS.T.to_dict()
        UR_LCH = (UR+LR+SR)*LCH #(7/6*UR)*LCH

        dep_cost_const = model.addConstr(dep_cost == gp.quicksum(pZ_k_dict[k]*PF*dtt_k[k]*ER*(3066 + 349.66*dt_k[k]) + pZ_k_dict[k]*PF*dt_k_last[k]*ER*(3066 + 349.66*dt_k_last[k]) for k in Z))
        log_cost_const = model.addConstr(log_cost == gp.quicksum(fFP_ij[i,j]*(nT_ij[i,j]*dFP_ij_LCM_LCH_TS_dict[i][j] + UR_LCH*aFP_ij[i,j]) for i in F for j in P))
        #dep_cost_const2 = model.addConstr(dep_cost >= 1e6)
        #log_cost_const2 = model.addConstr(log_cost >= 1e3)
        #obj_cost_const_ub = model.addConstr(dep_cost+log_cost <= 5.1e8)
        #obj_cost_const_lb = model.addConstr(dep_cost+log_cost >= 1e4)

        model.setObjective(dep_cost + log_cost, gp.GRB.MINIMIZE)
        model.setParam('Presolve', 2) #to reduce the model size and tighten it further, if possible
        model.setParam('MIPFocus', 3)#3 focus on the bound, 2 focus on the optimality, 1 focus on the feasibility
        model.setParam('Cuts', 3)#Use value 0 to shut off cuts, 1 for moderate cut generation, 2 for aggressive cut generation, and 3 for very aggressive cut generation. The default -1 value chooses automatically.
        #model.setParam('InfProofCuts',2)
        #model.setParam('MIPSepCuts',2)#1 for moderate cut generation, or 2 for aggressive cut generation
        #model.setParam('BQPCuts', 2)#Boolean Quadric Polytope (BQP) 1 for moderate cut generation, or 2 for aggressive cut generation
        #model.setParam('PreSparsify',1)
        model.setParam('IntegralityFocus',1)
        model.setParam('VarBranch', 2) #Maximum Infeasibility Branching (2), and Strong Branching (3)
        #model.setParam('ZeroHalfCuts', 2)# 1 for moderate cut generation, or 2 for aggressive cut generation
        
        #model.setParam('ImproveStartTime', 120)
        #model.setParam('NodeMethod',1)
        #model.setParam('NLPHeur', 0)
        #model.setParam('Heuristics', 1)
        #model.setParam('NoRelHeurTime', 120)#43200
        #model.setParam('DegenMoves', 100) #to avoid exhausts memory
        #model.setParam('Cutpasses', 100) #to avoid exhausts memory
        #model.setParam('GomoryPasses', 0) #to set the number of Gomory cut passes
        #model.setParam('StrongCGCuts', 2) #Strong cuts

        #model.setParam('NodefileStart', 0.5) #to avoid exhausts memory
        #model.setParam('Threads', 30) #to avoid exhausts memory
        #model.setParam('PreSparsify', 1) #This reduction can sometimes significantly reduce the number of non-zero values in the presolved model
        #model.setParam('CliqueCuts', 2)#Clique cuts
        #model.setParam('ScaleFlag', 1) #to scale the model
        #model.setParam('NonConvex', 2)#Strategy for handling non-convex quadratic programs
        model.optimize()
        #model.tune()
        #model.computeIIS()
        #model.write("model.ilp")
        #model.write('FoodBank_Final.lp')
        # Check the status and get the solution
        if model.status == gp.GRB.OPTIMAL:
            model.write(f'FoodBank_{keyP}-{keyZ}-PF005_NoLineal_Final.sol')
            print(f'Objective Value: {model.objVal}')
            print(f'MIP Gap {keyP}-{keyZ}: {model.MIPGap}')
        
        FP_solution, FP_variables, PZ_solution, PZ_variables, FP_noSol, FP_varNoSol = get_coordinates(F, P, Z, x_ij, y_jk, FBLocation_i, PODLocation_j, ZoneLocation_k)
        get_map(FBLocation_i, FP_solution, FP_variables, PZ_solution, PZ_variables, FP_noSol, FP_varNoSol, keyP, keyZ)
```

```python
import gurobipy as gp

# **Sets**:
nF = 1
F = np.arange(1, nF+1) #(Food Banks) i
nP = 228
nZ = 96

# **Parameters**:
pZ = pd.read_excel(pd.ExcelFile(data_path), sheet_name='ZIP_CODE_POPULATION-DEMAND', header=None)
pZ_k = pZ.iloc[1:,1:2].astype(float) # Population in Zone $Z_k$ (People) pZ_k.loc[1].values[0]
pZ_k_dict = pZ_k[1].to_dict()

dPZ_file = pd.read_excel(pd.ExcelFile(data_path), sheet_name='DISTANCE_POD_ZIP', index_col=0)
dPZ_jk = dPZ_file.astype(float) # Distance from $Z_k$ centroid to $P_j$ (Miles) dZP_kj.loc[1,1]
dPZ_jk_dict = dPZ_jk.T.to_dict()

dFP = pd.read_excel(pd.ExcelFile(data_path), sheet_name='DISTANCE_FOOD_BANK-PODS', index_col=0)
dFP_ij = dFP.astype(float) #Distance from $F_i$ to $P_j$ (Miles)
dFP_ij_dict = dFP_ij.T.to_dict()

cP = pd.read_excel(pd.ExcelFile(data_path), sheet_name='CAPACITY_POD', index_col=0)
cP_j = cP.iloc[0:,0:1].astype(float) # POD $P_j$ capacity (Pounds)
cP_j_dict = cP_j['Capacity_pounds_of_food_per_day'].to_dict()

# Number of Trucks availables
NT = nP*100
NT_i = pd.DataFrame([NT]) 
NT_i.columns = NT_i.columns + 1 
NT_i.index = NT_i.index + 1
NT_i_dict = NT_i[1].to_dict()

LCM = 0.969 #0.58 # Logistic Cost per Mile ($/Mile)
LCH = 71.78 # Logistic Cost per Hour ($/Hour)
AFP = 13.39/24 # Amount of Food required by Person (Pounds/Hour)

TC = 10000 # Trucks' Capacity (Pounds)
TS = 30 #12.5 # Trucks' Speed (Miles/Hour)
WS = 12.5 #3.1 # Walking Speed (Miles/Hour)
UR = 0.00005 # Unloading Rate (Hours/Pound)
SR = 0.0003 # Setting-up Rate (Hours/Pound)
LR = 0.00005 # Loading Rate (Hours/Pound)

PF = 0.05#0.1 # 0.005#% of People requiring Food
T = 10 # Period planning time (Days)

# Food Bank $F_i$ capacity (Pounds) Big-M
CF = np.sum(AFP*PF*pZ_k, axis=0)*T*24#*100
cF_i = pd.DataFrame([CF])
cF_i.index = cF_i.index + 1
cF_i_dict = cF_i[1].to_dict()

# Walking time (hours) from $Z_k$ to $P_j$
wt_jk = dPZ_jk/WS
wt_jk_dict = wt_jk.T.to_dict()
# Transportation time (hours) from $F_i$ to $P_j$
tt_ij = dFP_ij/TS
tt_ij_dict = tt_ij.T.to_dict()

M_Z = AFP*PF*np.max(pZ_k)*T*24*100
M_T = T*24*100
T_hours = T*24
AFP_PF_pZ_k = AFP*PF*pZ_k
AFP_PF_pZ_k_dict = AFP_PF_pZ_k[1].to_dict()

#PODs = {'PSN': P_small_near, 'PSF': P_small_far}
#Zones = {'ZSN': Z_small_near, 'ZSF': Z_small_far}
PODs = {'PMN': P_medium_near, 'PMF': P_medium_far}
Zones = {'ZMN': Z_medium_near, 'ZMF': Z_medium_far}
#nP = 228
#nZ = 96
#PODs = {'FULL': np.arange(1, nP+1)}
#Zones = {'NETWORK': np.arange(1, nZ+1)}

for keyP, P in PODs.items():
    for keyZ, Z in Zones.items():
        # ##MODEL##
        model = gp.Model('Food_Bank') 

        # ##DECISION VARIABLES##:
        # x_ij if $F_i$ sends food to $P_j$
        x_ij = {(i, j): model.addVar(vtype = gp.GRB.BINARY, name = "x_%d_%d" %(i, j)) for i in F for j in P}
        # y_kj if $Z_k$ is served by $P_j$  
        y_jk = {(j, k): model.addVar(vtype = gp.GRB.BINARY, name = "y_%d_%d" %(j, k)) for j in P for k in Z}

        # Amount of food (pounds) sent from $F_i$ to $P_j$
        aFP_ij = {(i, j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "aFP_%d_%d" %(i, j)) for i in F for j in P} 
        # Amount of food (pounds) that $P_j$ can serve to $Z_k$
        aPZ_jk = {(j, k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "aPZ_%d_%d" %(j, k)) for k in Z for j in P}
        # Number of trucks sent from $F_i$ to $P_j$
        nT_ij = {(i, j): model.addVar(vtype = gp.GRB.INTEGER, name = "nT_%d_%d" %(i, j)) for i in F for j in P} 
        # Order received time (hours) in $P_j$
        ort_ij = {(i,j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "ort_%d_%d" %(i,j)) for i in F for j in P}
        # Order placed time (hours) in $P_j$
        opt_ij = {(i,j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "opt_%d_%d" %(i,j)) for i in F for j in P}
        # Deprivation time (hours) that people in $Z_k$ wait while food is available in $P_j$. Por proporcion de la poblacion servida en ese POD.
        dt_jk = {(j, k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_%d_%d" %(j, k)) for j in P for k in Z}
        # Unloading time (hours) in $P_j$ related to the amount of food received from $F_i$
        ut_ij = {(i, j): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "ut_%d_%d" %(i, j)) for i in F for j in P}
        # Delivery frequency of trucks sent from $F_i$ to $P_j$
        fFP_ij = {(i, j): model.addVar(vtype = gp.GRB.INTEGER, name = "fFP_%d_%d" %(i, j)) for i in F for j in P}

        dt_k = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_%d" %(k)) for k in Z}
        dtt_k = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dtt_%d" %(k)) for k in Z}
        dt_k_last = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_last_%d" %(k)) for k in Z}
        dt_k_start = {(k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dt_start_%d" %(k)) for k in Z}
        ort_jk = {(j,k): model.addVar(vtype = gp.GRB.CONTINUOUS, name = "ort_%d_%d" %(j,k)) for j in P for k in Z}

        dep_cost = model.addVar(vtype = gp.GRB.CONTINUOUS, name = "dep_cost")
        log_cost = model.addVar(vtype = gp.GRB.CONTINUOUS, name = "log_cost")
        
        #Update model (it actully puts everything togeter)
        model.update()

        # ##CONSTRAINTS##:

        Fi_sendsFood_Pj = {model.addConstr(gp.quicksum(x_ij[i,j] for j in P) >= 1, name=f'F{i}_sendsFood_Pj') for i in F}
        Pj_receiveFood_Fi = {model.addConstr(gp.quicksum(x_ij[i,j] for i in F) <= 1, name=f'P{j}_receiveFood_Fi') for j in P}
        Pj_sendsFood_Zk = {model.addConstr(gp.quicksum(y_jk[j,k] for k in Z) <= 1, name=f'P{j}_sendsFood_Zk') for j in P}
        Zk_receiveFood_Pj = {model.addConstr(gp.quicksum(y_jk[j,k] for j in P) >= 1, name=f'Z{k}_receiveFood_Pj') for k in Z}

        Fi_sendsFood_Pj_TruckCapacity = {model.addConstr(aFP_ij[i,j] <= x_ij[i,j]*cP_j_dict[j]) for i in F for j in P}
        FoodBankCapacity = {model.addConstr(gp.quicksum(aFP_ij[i,j] for j in P) <= cF_i_dict[i]) for i in F}
        Truck_available = {model.addConstr(gp.quicksum(nT_ij[i,j] for j in P) <= NT_i_dict[i]) for i in F}

        FoodGet_FoodSent = {model.addConstr(gp.quicksum(aFP_ij[i,j] for i in F) == gp.quicksum(aPZ_jk[j,k] for k in Z)) for j in P} 
        Food_sent_capacity = {model.addConstr(nT_ij[i,j]*TC >= aFP_ij[i,j]) for i in F for j in P}

        UnloadingTime = {model.addConstr(ut_ij[i, j] == (UR+SR)*aFP_ij[i, j]) for i in F for j in P}

        OrderPlacedTime = {model.addConstr(opt_ij[i,j] + 2*tt_ij_dict[i][j]*x_ij[i,j] + LR*ut_ij[i,j] == ort_ij[i,j]) for i in F for j in P}

        FoodDemand = {model.addConstr(gp.quicksum(aPZ_jk[j,k]*fFP_ij[i,j] for i in F for j in P) <= AFP_PF_pZ_k_dict[k]*T_hours) for k in Z}
        DeprivationTime = {model.addConstr(AFP_PF_pZ_k_dict[k]*ort_ij[i,j]*y_jk[j,k] == AFP_PF_pZ_k_dict[k]*dt_jk[j,k] + aPZ_jk[j,k]) for i in F for j in P for k in Z }

        #DeprivationTime2 = {model.addConstr(dt_jk[j,k] >= ut_ij[i,j]*y_jk[j,k] + wt_jk_dict[j][k]*y_jk[j,k]) for i in F for j in P for k in Z }
        DeprivationTimeStart = {model.addConstr(dt_k_start[k] >= gp.quicksum(ut_ij[i,j]*y_jk[j,k] for i in F for j in P) + wt_jk_dict[j][k]*y_jk[j,k]) for j in P for k in Z }

        PlaningTime = {model.addConstr(fFP_ij[i,j]*ort_ij[i,j] <= T_hours*x_ij[i,j]) for i in F for j in P}

        M_T = T_hours
        #new0 = {model.addConstr(dt_jk[j,k] <= M_T*y_jk[j,k]) for j in P for k in Z }
        new0 = {model.addConstr(dt_jk[j,k] <= ort_ij[i,j]*y_jk[j,k]) for i in F for j in P for k in Z }
        new1 = {model.addConstr(aPZ_jk[j,k] <= cP_j_dict[j]*y_jk[j,k]) for j in P for k in Z}
        new2 = {model.addConstr(x_ij[i,j] <= gp.quicksum(y_jk[j,k] for k in Z)) for i in F for j in P}
        #new3 = {model.addConstr(ut_ij[i,j] <= M_T*x_ij[i,j]) for i in F for j in P}
        new3 = {model.addConstr(ut_ij[i,j] <= ort_ij[i,j]) for i in F for j in P}
        new4 = {model.addConstr(dt_k_last[k] <= M_T) for k in Z}
        new41 = {model.addConstr(dt_k_start[k] <= M_T) for k in Z}
        new5 = {model.addConstr(dt_k[k] <= M_T) for k in Z}
        new7 = {model.addConstr(ort_ij[i,j] <= M_T*x_ij[i,j]) for i in F for j in P}
        new8 = {model.addConstr(fFP_ij[i,j] <= 10*M_T*nT_ij[i,j]) for i in F for j in P}
        neww81 = {model.addConstr(fFP_ij[i,j] >= x_ij[i,j]) for i in F for j in P}
        #new9 = {model.addConstr(opt_ij[i,j] <= M_T*x_ij[i,j]) for i in F for j in P}
        new9 = {model.addConstr(opt_ij[i,j] <= ort_ij[i,j]) for i in F for j in P}
        new10 = {model.addConstr(nT_ij[i,j] <= NT_i_dict[i]*x_ij[i,j]) for i in F for j in P}#NT_i_dict[i]

        ort_jk_const = {model.addConstr(ort_jk[j,k] == gp.quicksum(ort_ij[i,j]*y_jk[j,k] for i in F)) for j in P for k in Z}
        #DeprivationTimeFinal = {model.addConstr(dt_k_last[k] >= T_hours*y_jk[j,k] - fFP_ij[i,j]*ort_ij[i,j]) for i in F for j in P for k in Z}
        DeprivationTimeFinal = {model.addConstr(AFP_PF_pZ_k_dict[k]*dt_k_last[k] >=  AFP_PF_pZ_k_dict[k]*T_hours*y_jk[j,k] -  AFP_PF_pZ_k_dict[k]*fFP_ij[i,j]*ort_jk[j,k] -  AFP_PF_pZ_k_dict[k]*dt_jk[j,k] - aPZ_jk[j,k]) for i in F for j in P for k in Z}
        dt_k_total = {model.addConstr(dt_k[k] >= dt_jk[j,k]) for j in P for k in Z}
        dtt_k_total = {model.addConstr(dtt_k[k] >= dt_k[k]*fFP_ij[i,j]) for i in F for j in P for k in Z}

        # ##OBJECTIVE FUNCTION##:
        ER = 1/2000 # Exchange Rate USD to COP Base on the paper Discrete choice Cantillo et al. (2018)
        dFP_ij_LCM_LCH_TS = 2*dFP_ij*(LCM+(LCH/TS))
        dFP_ij_LCM_LCH_TS_dict = dFP_ij_LCM_LCH_TS.T.to_dict()
        UR_LCH = (UR+LR+SR)*LCH #(7/6*UR)*LCH

        dep_cost_const = model.addConstr(dep_cost == gp.quicksum(pZ_k_dict[k]*PF*dtt_k[k]*ER*(3066 + 349.66*dt_k[k]) + pZ_k_dict[k]*PF*dt_k_start[k]*ER*(3066 + 349.66*dt_k_start[k]) + pZ_k_dict[k]*PF*dt_k_last[k]*ER*(3066 + 349.66*dt_k_last[k]) for k in Z))
        log_cost_const = model.addConstr(log_cost == gp.quicksum(fFP_ij[i,j]*(nT_ij[i,j]*dFP_ij_LCM_LCH_TS_dict[i][j] + UR_LCH*aFP_ij[i,j]) for i in F for j in P))
        #dep_cost_const2 = model.addConstr(dep_cost >= 1e6)
        #log_cost_const2 = model.addConstr(log_cost >= 1e3)
        #obj_cost_const_ub = model.addConstr(dep_cost+log_cost <= 5.1e8)
        #obj_cost_const_lb = model.addConstr(dep_cost+log_cost >= 1e4)

        model.setObjective(dep_cost + log_cost, gp.GRB.MINIMIZE)
        model.setParam('Presolve', 2) #to reduce the model size and tighten it further, if possible
        #model.setParam('MIPFocus', 3)#3 focus on the bound, 2 focus on the optimality, 1 focus on the feasibility
        #model.setParam('Cuts', 0)#Use value 0 to shut off cuts, 1 for moderate cut generation, 2 for aggressive cut generation, and 3 for very aggressive cut generation. The default -1 value chooses automatically.
        #model.setParam('InfProofCuts',2)
        #model.setParam('MIPSepCuts',2)#1 for moderate cut generation, or 2 for aggressive cut generation
        #model.setParam('BQPCuts', 2)#Boolean Quadric Polytope (BQP) 1 for moderate cut generation, or 2 for aggressive cut generation
        #model.setParam('PreSparsify',1)
        model.setParam('IntegralityFocus',1)
        #model.setParam('VarBranch', 2) #Maximum Infeasibility Branching (2), and Strong Branching (3)
        #model.setParam('ZeroHalfCuts', 2)# 1 for moderate cut generation, or 2 for aggressive cut generation
        
        #model.setParam('ImproveStartTime', 120)
        #model.setParam('NodeMethod',1)
        #model.setParam('NLPHeur', 0)
        #model.setParam('Heuristics', 1)
        #model.setParam('NoRelHeurTime', 120)#43200
        #model.setParam('DegenMoves', 100) #to avoid exhausts memory
        #model.setParam('Cutpasses', 100) #to avoid exhausts memory
        #model.setParam('GomoryPasses', 0) #to set the number of Gomory cut passes
        #model.setParam('StrongCGCuts', 2) #Strong cuts

        #model.setParam('NodefileStart', 0.5) #to avoid exhausts memory
        #model.setParam('Threads', 30) #to avoid exhausts memory
        #model.setParam('PreSparsify', 1) #This reduction can sometimes significantly reduce the number of non-zero values in the presolved model
        #model.setParam('CliqueCuts', 2)#Clique cuts
        #model.setParam('ScaleFlag', 1) #to scale the model
        #model.setParam('NonConvex', 2)#Strategy for handling non-convex quadratic programs
        model.optimize()
        #model.tune()
        #model.computeIIS()
        #model.write("model.ilp")
        #model.write('FoodBank_Final.lp')
        # Check the status and get the solution
        if model.status == gp.GRB.OPTIMAL:
            model.write(f'FoodBank_{keyP}-{keyZ}-PF005_NoLineal_Final.sol')
            print(f'Objective Value: {model.objVal}')
            print(f'MIP Gap {keyP}-{keyZ}: {model.MIPGap}')
        
        FP_solution, FP_variables, PZ_solution, PZ_variables, FP_noSol, FP_varNoSol = get_coordinates(F, P, Z, x_ij, y_jk, FBLocation_i, PODLocation_j, ZoneLocation_k)
        get_map(FBLocation_i, FP_solution, FP_variables, PZ_solution, PZ_variables, FP_noSol, FP_varNoSol, keyP, keyZ)
```

```python
def read_gurobi_solution(file_path):
    # Step 1: Read the file and skip the first line
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Step 2: Extract the objective value from the second line and remove the '#' symbol
    objective_line = lines[1]
    objective_value = float(objective_line.split('=')[1].strip())

    # Remove the first two lines
    lines = lines[2:]

    # Step 3: Write the remaining lines to a new file or directly to a DataFrame
    temp_file_path = 'temp_file.txt'
    with open(temp_file_path, 'w') as temp_file:
        temp_file.writelines(lines)

    # Step 4: Read the cleaned data into a DataFrame
    df = pd.read_csv(temp_file_path, delim_whitespace=True, header=None, names=['Variable', 'Value'])

    # Optional: Delete the temporary file if needed
    import os
    os.remove(temp_file_path)

    # Step 5: Add the objective value to the DataFrame
    df_objective = pd.DataFrame([['Objective_value', objective_value]], columns=['Variable', 'Value'])
    df = pd.concat([df_objective, df], ignore_index=True)
    return df
```
